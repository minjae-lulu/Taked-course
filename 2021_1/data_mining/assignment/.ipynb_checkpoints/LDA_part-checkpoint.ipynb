$\textbf{1_1. Principal Component Analysis & Linear Discriminant Analysis}$

$\textbf{Principal Component Analysis}$ : Principal Component Analysis(PCA) is the most well-known and widely used technique to resolve the multicollinearity issue. When PCA projects data into a vector, it looks for vectors with large variances in the projections. It reduces the dimension of the data by eliminating vectors that cause less variance in projections. PCA aim is to reduce high-dimensional data to low-dimensional, enabling visualization or eliminating noise.

$\textbf{Linear Discriminant Analysis}$ : Linear Discriminant Analysis's(LDA) objective is to perform dimensionality reduction while preserving ans much of the class discriminatory information as possible. LDA finds vectors with two properties. One is a vector that allows the sum of the distances between the means of the projections belonging to each class to be maximum when the data points are projected. Another is a vector that allows the variance of projections within a class to be minimal when projecting data points. So, LDA aim is feature extraction that the data is best separable.
    
$\textbf{Difference}$ : PCA is unsupervised learning that doesn't has label, but LDA is supervised learning that has label. Both PCA and LDA use a technique eigenvalue decomposition and projection. But, PCA projects data into vectors that widen the variance of the data to better understand it and LDA projects data into vectors that allow distances between classes and small variances within classes to classify them according to classes. Also, PCA reshape the new independent variables by using linear combination, but LDA does not reshape the new independent variable by using linear combination.

<br>

$\textbf{1_2. Decision Tree & Random Forest}$

$\textbf{Decision Tree}$ : Decision trees are analytical methods that represent decision-making rules in a tree structure to classify the entire data into several small groups or make predictions. Decision tree is supervised learning. Decision trees are classified as classification trees if the target variable is discrete and regression trees if the target variable is continuous. So, decision tree can use both classification and regression. Decision tree has many advantages. For example, easy to interpret, simple and quick to fit, handle categorical predictors naturally and so on. But has disadvantages that accuracy, instability and overfitting. Decision tree uses two methods: pre-pruning and post-pruning to address overfitting.

$\textbf{Random Forest}$ : We use random forest to reduce decision tree's disadvantage that accuracy and instability. So, random forest is also supervised learning. Random forests go through a process called bagging that chose random with replacement from the data. And then sum all these replaced data and averaged. With these features, random forests avoid overfitting by reducing the influence of trees where overfitting appears.
    
$\textbf{Difference}$ : Random forest is designed to solve the problem of decision trees, so there are many similarities, but some feature are different. While decision trees use one model, random forests use subset of decision trees model that allow replacement. Therefore, random forests have relatively better accuracy than decision trees. Also, random forests do not occur overfitting than decision trees. Because of its high performance, most use random forest, but some appropriate situation use decisioin tree. The reason is random forest takes time using a lot of memory, while decision trees are simple and fast.

```python
import matplotlib.pyplot as plt
TPR = [1,1,1,0.83,0.83,0.67,0.5,0.5,0.33,0.17,0]
FPR = [1,0.75,0.5,0.5,0.25,0.25,0.25,0,0,0,0]

for i in range(0,10):
    TPR[i] *= 100
    FPR[i] *= 100
    
```

```python
plt.plot(FPR,TPR,'k')
plt.scatter(FPR, TPR)
plt.xlabel("FPR")
plt.ylabel("TPR")
plt.title("ROC curve")
```

ROC curve has x_axis that means the ratio of FPR and y_axis that means the ratio of TPR. Because ROC is a graph, it is difficult to compare with clear figures. So we use the area value under the graph, which is AUC(Area Under Curve). This area's max value is 1, and the closer to 1, the better the model. Closer 1 means that the TPR value is larger than the FPR (graph line is close left up)
